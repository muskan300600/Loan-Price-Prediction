```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("train_u6lujuX_CVtuZ9i (1).csv")
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.replace({"Loan_Status":{'N':0,'Y':1}})
```

```python
data.corr()
```

```python
data.Gender.value_counts()
```

```python
data.Gender.fillna('-1',inplace=True)
data
```

```python
data.isnull().sum()
```

```python
data.Married.fillna('-1',inplace=True)
data.isnull().sum()
```

```python
data.Self_Employed.fillna('-1',inplace=True)
data.isnull().sum()
```

```python
data.Dependents.value_counts()
```

```python
data.Dependents.fillna('-1',inplace=True)
data.isnull().sum()
```

```python
mean_la = data.LoanAmount.mean()
data.LoanAmount.fillna(mean_la,inplace=True)
data.isnull().sum()
```

```python
median_lat = data.Loan_Amount_Term.median()
median_lat
data.Loan_Amount_Term.fillna(median_lat,inplace=True)
data.isnull().sum()
```

```python
median_ch = data.Credit_History.median()
data.Credit_History.fillna(median_ch,inplace=True)
data.isnull().sum()
```

```python
Gen = pd.get_dummies(data.Gender,drop_first=True)
Marry = pd.get_dummies(data.Married,drop_first=True)
Edu = pd.get_dummies(data.Education,drop_first=True)
Emp = pd.get_dummies(data.Self_Employed,drop_first=True)
Prop = pd.get_dummies(data.Property_Area,drop_first=True)
```

```python
data = pd.concat([data,Gen,Marry,Edu,Emp,Prop],axis=1)
data
```

```python
data.corr()
```

```python
data=data.drop(['Gender','Married','Education','Self_Employed','Property_Area'],axis=1)
data
```

```python
data.corr()
```

```python
data.isnull().sum()
```

```python
data=data.drop(['Female','No','No'],axis=1)
data
```

```python
data.Dependents = data.Dependents.replace(to_replace='3+',value=4)
data
```

```python
X = data.drop(['Loan_Status','Loan_ID'],axis=1)
y=data.Loan_Status
X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,stratify=y,random_state=0)
```

```python
X_train.shape,y_train.shape
```

```python
model = svm.SVC(gamma='auto',C=1,kernel='linear')
model.fit(X_train,y_train)
```

```python
training = model.predict(X_train)
score1 = accuracy_score(training,y_train)
score1
```

```python
training2 = model.predict(X_test)
score2 = accuracy_score(training2,y_test)
score2
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred1= lr.predict(X_train)
score3= accuracy_score(pred1,y_train)
score3
```

```python
pred2= lr.predict(X_test)
score4= accuracy_score(pred2,y_test)
score4
```
